```python
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from PIL import Image
import torch
```

```python
image_path = "Hero-1di.png"  # Replace with your image path
image = Image.open(image_path)
```

```python
# Define the transform
transform = transforms.ToTensor()

# Apply transformation
tensor_image = transform(image)
```

```python
tensor_image.shape
```

```python
# Display Original and Transformed Image
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Original Image
ax[0].imshow(image)
ax[0].set_title("Original Image (PIL)")
ax[0].axis("off")

# Transformed Tensor Image (Converted to NumPy for Display)
ax[1].imshow(tensor_image.permute(1, 2, 0))  # Convert (C, H, W) to (H, W, C)
ax[1].set_title("After ToTensor() (Tensor)")
ax[1].axis("off")

plt.show()
```
